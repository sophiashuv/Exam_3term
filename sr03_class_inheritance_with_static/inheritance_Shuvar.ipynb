```python
from datetime import datetime


class Validation:
    """Class for Validation representation."""

    @staticmethod
    def validateStr(func):
        def validateStrWrapper(w, value):
            if not(isinstance(value, str)):
                raise TypeError('must be an str')
            return func(w, value)
        return validateStrWrapper

    @staticmethod
    def validateFloat(func):
        def validateFloatWrapper(w, value):
            if not(isinstance(value, float)) and not(isinstance(value, int)):
                raise TypeError('must be an float')
            return func(w, value)
        return validateFloatWrapper

    @staticmethod
    def validateInt(func):
        def validateIntWrapper(w, value):
            if not(isinstance(value, int)):
                raise TypeError('must be an int')
            return func(w, value)
        return validateIntWrapper

    @staticmethod
    def validateSize(func):
        def validateSizeWrapper(w, value):
            if not(isinstance(value, int)) and not(isinstance(value, int)):
                raise TypeError('must be an int')
            if value < 0:
                raise ValueError("mast be positive")
            return func(w, value)
        return validateSizeWrapper

    @staticmethod
    def validateAmount(func):
        def validateAmountWrapper(w, value):
            if not(isinstance(value, int)):
                raise TypeError('must be an int')
            if value < 0:
                raise ValueError("mast be positive")
            return func(w, value)
        return validateSizeWrapper

    @staticmethod
    def validateDate(func):
        def validateDateWrapper(w, date):
            datetime.strptime(date, '%Y-%m-%d')
            return func(w, date)
        return validateDateWrapper


    @staticmethod
    def validateFileName(end=".txt"):
        def validateFileNameDecorator(func):
            def validateFileNameWrapper(L, filename):
                if not filename.endswith(end):
                    raise ValueError("Incorrect filename, should end with ." + end + ".")
                return func(L, filename)
            return validateFileNameWrapper
        return validateFileNameDecorator
```

## task 1

```python
class Delivery:
    def __init__(self, reciever_country, sender_country, main_price):
        self.reciever_country = reciever_country
        self.sender_country = sender_country
        self.main_price = main_price

    @property
    def reciever_country(self):
        return self._reciever_country

    @reciever_country.setter
    @Validation.validateStr
    def reciever_country(self, val):
       self._reciever_country = val
    
    @property
    def sender_country(self):
        return self._sender_country

    @sender_country.setter
    @Validation.validateStr
    def sender_country(self, val):
        if val == "":
            self._sender_country = "Sender Country"
        else:
            self._sender_country = val

    @property
    def main_price(self):
        return self._main_price

    @main_price.setter
    @Validation.validateFloat
    def main_price(self, val):
        self._main_price = val

    def __str__(self):
        return f'reciever country = {self.reciever_country}, sender country = {self.sender_country}, main price = {self.main_price}'

    def __repr__(self):
        return f'reciever_country = {self.reciever_country}, sender_country = {self.sender_country}, main_price = {self.main_price}'
```

```python
deliveries = [Delivery("Afganistan", "Azibajan", 520), 
Delivery("France", "Turkey", 7452), 
Delivery("Germany", "Colombia", 410), 
Delivery("USA", "China", 10001), 
Delivery("Poland", "Belarus", 100)]

for delivery in deliveries:
    print(delivery)
```

```python
maximal_price_deliveries = max(deliveries, key=lambda d: d.main_price)
print(f'<{maximal_price_deliveries.reciever_country}, {maximal_price_deliveries.sender_country}>')
```

## Task 2

```python
class Express(Delivery):
    def __init__(self, reciever_country, sender_country, main_price, commision=15):
        super().__init__(reciever_country, sender_country, main_price)
        self.commision = commision

    @property
    def commision(self):
        return self._commision

    @commision.setter
    @Validation.validateInt
    def commision(self, val):
        self._commision = val

    def express_price(self):
        return self.main_price * (1 + self.commision/100)

    def __str__(self):
        return super().__str__() + f', commision = {self.commision}'

    def __repr__(self):
        return super().__repr__() + f', commision = {self.commision}'
```

```python
expresses = [Express("Urugway", "PipiPupu", 452, 10), 
Express("Kalinka", "Malinka", 8752), 
Express("TikTok", "PitStop", 142), 
Express("Inigini", "Denifinki", 2), 
Express("Lola", "Pola", 8752)]

for express in expresses:
    print(express)
```

```python
max_price_express = max(expresses, key=lambda p: p.express_price())
maximal_price_expresses = list(filter(lambda p: max_price_express.express_price() == p.express_price(), expresses))

for el in maximal_price_expresses:
    print(f'<{el.reciever_country}, {el.sender_country}>: ${el.express_price()}')
```

## Task 3 

```python
class Banderol(Delivery):
    def __init__(self, reciever_country, sender_country, main_price, weight, k=10):
        super().__init__(reciever_country, sender_country, main_price)
        self.weight = weight
        self.k = k

    @property
    def weight(self):
        return self._weight

    @weight.setter
    @Validation.validateSize
    def weight(self, val):
        self._weight = val

    @property
    def k(self):
        return self._k

    @k.setter
    @Validation.validateSize
    def k(self, val):
        self._k = val

    def parcel_price(self):
        return self.main_price + self.weight * self.k/100

    def __str__(self):
        return super().__str__() + f', weight = {self.weight}, k = {self.k}'

    def __repr__(self):
        return super().__repr__() + f', weight = {self.weight}, k = {self.k}'
```

```python
class Parcel(Banderol):
    def __init__(self, reciever_country, sender_country, main_price, weight, volume, q=13):
        super().__init__(reciever_country, sender_country, main_price, weight)
        self.volume = volume
        self.q = q

    @property
    def volume(self):
        return self._volume

    @volume.setter
    @Validation.validateSize
    def volume(self, val):
        self._volume = val

    @property
    def q(self):
        return self._q

    @q.setter
    @Validation.validateSize
    def q(self, val):
        self._q = val

    def parcel_price(self):
        return super().parcel_price() + self.volume * self.q/100

    def __str__(self):
        return super().__str__() + f', volume = {self.volume}, q = {self.q}'

    def __repr__(self):
        return super().__repr__() + f', volume = {self.volume}, q = {self.q}'
```

```python
ban_par = [Banderol("Arial", "Infi", 1500, 14), 
Banderol("Helsinki", "Parapaparam", 103, 28), 
Banderol("Jiki", "Siki", 7800, 2), 
Parcel("Tutu", "Tata", 1478, 54, 68), 
Parcel("Seka", "Neka", 7412, 789, 98), 
Parcel("Lika", "Dika", 1458, 125, 245)]

for i in ban_par:
    print(f'<{i.reciever_country} , {i.sender_country}> : {i.parcel_price()}')
```

```python
max_price_parcel = max(ban_par, key=lambda p: p.parcel_price())
maximal_price_parcel = list(filter(lambda p: max_price_parcel.parcel_price() == p.parcel_price(), ban_par))

for el in maximal_price_parcel:
    print(f'<{el.reciever_country}, {el.sender_country}>: ${el.parcel_price()}')
```
